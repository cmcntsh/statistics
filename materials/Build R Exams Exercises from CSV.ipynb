```python
# set up template strings for each question type

# True False questions
def TFstring(title,question,TF1,TF2,TF3,TF4):
    
    TFstring = (
    f"Question\n"
    f"========\n"
    f"{question}\n"
    f"\n"
    f"Answerlist\n"
    f"----------\n"
    f"* True\n"
    f"* False\n"
    f"\n"
    f"\n"
    f"Solution\n"
    f"========\n"
    f"\n"
    f"Answerlist\n"
    f"----------\n"
    f"* {TF1}.\n"
    f"* {TF2}.\n"
    f"\n"
    f"\n"
    f"Meta-information\n"
    f"================\n"
    f"exname: {title}\n"
    f"extype: mchoice\n"
    f"exsolution: {TF3}{TF4}\n"
    f"exshuffle: TRUE"
    )
    return TFstring

# Multiple Choice questions
def MCstring(title,question,correct,answers):
    
    MCstring = (
    f"Question\n"
    f"========\n"
    f"{question}\n"
    f"\n"
    f"Answerlist\n"
    f"----------\n"        
    )
    for answer in answers:
        MCstring += f"* " + answer +"\n"
    
    MCstring += (
    f"\n"
    f"\n"
    f"Solution\n"
    f"========\n"
    f"\n"
    f"Answerlist\n"
    f"----------\n"
    )
    i=1
    for answer in answers:
        if str(correct)==str(i):
            MCstring += f"* True.\n"
        else:
            MCstring += f"* False.\n"
        i+=1
        
    MCstring += (
    f"\n"
    f"\n"
    f"Meta-information\n"
    f"================\n"
    f"exname: {title}\n"
    f"extype: mchoice\n"
    f"exsolution: "
    )
    i=1
    for answer in answers:
        if str(correct)==str(i):
            MCstring += '1'
        else:
            MCstring += '0'
        i+=1
    
    MCstring += (
    f"\n"
    f"exshuffle: TRUE"
    
    )
    
    return MCstring

# Multiple Response questions
def MRstring(title,question,correct,answers):
    
    MRstring = (
    f"Question\n"
    f"========\n"
    f"{question}\n"
    f"\n"
    f"Answerlist\n"
    f"----------\n"        
    )
    for answer in answers:
        MRstring += f"* " + answer +"\n"
    
    MRstring += (
    f"\n"
    f"\n"
    f"Solution\n"
    f"========\n"
    f"\n"
    f"Answerlist\n"
    f"----------\n"
    )
    i=1
    for answer in answers:
        if str(i) in str(correct):
            MRstring += f"* True.\n"
        else:
            MRstring += f"* False.\n"
        i+=1
        
    MRstring += (
    f"\n"
    f"\n"
    f"Meta-information\n"
    f"================\n"
    f"exname: {title}\n"
    f"extype: mchoice\n"
    f"exsolution: "
    )
    i=1
    for answer in answers:
        if str(i) in str(correct):
            MRstring += '1'
        else:
            MRstring += '0'
        i+=1
    
    MRstring += (
    f"\n"
    f"exshuffle: TRUE"
    
    )
    
    return MRstring

print(TFstring)
```

```python
import csv
import re

# count lines in file
with open(file_path, encoding='utf-8-sig') as csv_obj:
    read_CSV = csv.reader(csv_obj, delimiter=',')
    row_count = sum(1 for row in read_CSV)
    print(str(row_count) + ' total rows in file.')

# iterate through rows in file
with open(file_path, encoding='utf-8-sig') as csv_obj:
    read_CSV = csv.reader(csv_obj, delimiter=',')
    print('file opened')
    
    # set up variables to count rows and types of questions
    # TF = True False questions
    # MC = Multiple Choice questions (select 1 answer choice)
    # MR = Multiple Response questions (select > 1 answer choice)
    line_count = 0
    TFcount = 0
    MCcount = 0
    MRcount = 0
    Othercount = 0
    
    # collect titles in a list
    title_list = []
    
    # sort titles by question format and subject (used by R exams for sampling questions)
    groups_dict = {
        'stats':[]
        ,'R':[]
        ,'valid':[]
        ,'vars':[]
        ,'LoM':[]        
    }
    
    
    
    for row in read_CSV:
        # first row will be a header. We can skip that.
        if line_count > 0:
            # load row information into variables
            qtype = row[0]
            title = row[1]
            points = row[2] 
            question = row[3]
            correct = row[4]
            answers = []
            
            # collect titles in a list
            title_list.append(title + '.Rmd')
            
            # sort titles by topic and append file name
            for key in groups_dict.keys():
                if key in title:
                    groups_dict[key].append(title + '.Rmd')
            
            for answer in row[5:14]:
                if answer>'':
                    answers.append(answer)
            
            R_exercise_file = f'{file_path[:end_path_pos]}/R exams exercises/{title}.Rmd'
            
            
            
            # check what the question type is in each row
            if qtype=="TF":
                if correct=="TRUE":
                    TF1 = 'True'
                    TF2 = 'False'
                    TF3 = '1'
                    TF4 = '0'
                    #print(TF1)
                elif correct=="FALSE":
                    TF1 = 'False'
                    TF2 = 'True'
                    TF3 = '0'
                    TF4 = '1'
                    #print(TF1)
                output_string = TFstring(title,question,TF1,TF2,TF3,TF4)
                with open(R_exercise_file, 'w') as f:
                    f.write(output_string)
                
                TFcount +=1
            
            elif qtype=="MC":
                output_string = MCstring(title,question,correct,answers)
                with open(R_exercise_file, 'w') as f:
                    f.write(output_string)
                MCcount +=1
            elif qtype=="MR":
                output_string = MRstring(title,question,correct,answers)
                with open(R_exercise_file, 'w') as f:
                    f.write(output_string)
                MRcount +=1
            else:
                Othercount +=1
            
        
        line_count +=1
        
    print("True False = " + str(TFcount))
    print("Multiple Choice = " + str(MCcount))
    print("Multiple Response = " + str(MRcount))
    print("Other = " + str(Othercount) + '\n')
    print('R exams .Rmd files written for questions in folder:' + R_exercise_file[:end_path_pos+19] + '\n')
    print('List of all files: ' + str(title_list) + '\n')
    print('List of all files sorted by topic: ' + str(groups_dict) + '\n')
    
    for key in groups_dict:
        print(str(key) + ' questions: ' + str(len(groups_dict[key])) + ' (' + str(round(len(groups_dict[key])/len(title_list)*100,1)) + '%)')
    

```
